```python
import numpy as np
from epidemic_models.NETWORK import *
import urllib.request
import pandas as pd
import datetime
from 
from scipy.optimize import curve_fit
```

```python
for radius in [0.25, 1, 2]:
    dt=0.03
    pt = 0.05
    mobility = 1
    gamma = 0.1
    N = 5
    #radius = 2
    total_population = 200

    sir = np.zeros((int(100/dt),3), dtype=float)
    trials = 10
    for m in range(trials):
        trial, population, locations = simulate1D(pt, mobility, gamma, N, radius, total_population, dt)
        sir += trial/trials

    #plotting network
    for pupil in population:
        for contact in pupil.contacts:
            friend = population[contact]
            plt.plot([pupil.x,friend.x],[pupil.y,friend.y],color='b',alpha=0.1)
    plt.scatter(locations[:,0],locations[:,1])
    plt.title('Network Connections')
    plt.show()
        
    # plot results
    sir = sir[:int(60/dt)]
    t = np.arange(len(sir[:,0]))*dt
    plt.plot(t, sir[:,0], label="S")
    plt.plot(t, sir[:,1], label="I")
    plt.plot(t, sir[:,2], label="R")
    plt.legend()
    plt.title('SIR Model')
    plt.xlabel('Weeks')
    plt.ylabel('Population')
    plt.show()
```
